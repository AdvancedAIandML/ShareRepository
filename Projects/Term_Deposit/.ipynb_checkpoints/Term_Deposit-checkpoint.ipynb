# Feature Engineering

```python
#Feature engineering for categorical features
dt=ds.copy()

dt.pdays[dt['pdays']==-1]=0

dt.default[dt['default']=='yes']=1
dt.default[dt['default']=='no']=0

dt.housing[dt['housing']=='yes']=1
dt.housing[dt['housing']=='no']=0

dt.loan[dt['loan']=='yes']=1
dt.loan[dt['loan']=='no']=0

dt.education[dt['education']=='unknown']=0
dt.education[dt['education']=='primary']=1
dt.education[dt['education']=='secondary']=2
dt.education[dt['education']=='tertiary']=3

dt.marital[dt['marital']=='single']=1
dt.marital[dt['marital']=='married']=2
dt.marital[dt['marital']=='divorced']=3

dt.month[dt['month']=='jan']=1
dt.month[dt['month']=='feb']=2
dt.month[dt['month']=='mar']=3
dt.month[dt['month']=='apr']=4
dt.month[dt['month']=='may']=5
dt.month[dt['month']=='jun']=6
dt.month[dt['month']=='jul']=7
dt.month[dt['month']=='aug']=8
dt.month[dt['month']=='sep']=9
dt.month[dt['month']=='oct']=10
dt.month[dt['month']=='nov']=11
dt.month[dt['month']=='dec']=12

dt.y[dt['y']=='yes']=1
dt.y[dt['y']=='no']=0


dt=pd.get_dummies(dt, columns = ['poutcome','contact','job'])


dt.head()
```

```python
# Get correlation between features

corrm=dt.corr()
corrm
```

```python
# Check for features which are highly correlated

# # Function to find highly correlated features:

def correlation(dataset, threshold):
    col_corr = set()  
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]  
                col_corr.add(colname)
    return col_corr

#Getting features which 70% correlated

correlation(dt,0.70)
```

Theres not much correlation observed between the features as shown above.

```python
# Visualize the correlation
plt.figure(figsize = (16,12))
sns.heatmap(corrm, annot=True)
```

```python
# Splitting Dataset into Training and Testing Datasets

sdn=dt.copy()

y=sdn['y']
y=y.astype('int')

x=sdn.drop('y',inplace=True,axis=1)
x=sdn

X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size =0.2)


# Scaling data for better accuracy with respect to models
scaler = StandardScaler()  
scaler.fit(X_train)

X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test)
```

# Model Building

```python
# Decision tree with depth = 8

model_8 = DecisionTreeClassifier(random_state=1, max_depth=8,criterion = "gini")
model_8.fit(X_train, Y_train)
model_8_score_train = model_8.score(X_train, Y_train)
print("Training score: ",model_8_score_train)
model_8_score_test = model_8.score(X_test, Y_test)
print("Testing score: ",model_8_score_test)
```

```python
# Performance and Accuracy Metrics for Decision Tree: Max Depth 8

pred_8 = model_8.predict(X_test)

#Construct the Confusion Matrix

labels=[0, 1]
cm = confusion_matrix(Y_test, pred_8, labels=[0, 1])
print(cm)
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
classes=["Success", "Default"]
cmap = plt.cm.Blues
plt.title('Confusion matrix')
fig.colorbar(cax)
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.imshow(cm, cmap=cmap)

# Precision, Recall, F1score

precision = precision_score(Y_test, pred_8)
print('Precision: %f' % precision)
# recall: tp / (tp + fn)
recall = recall_score(Y_test, pred_8)
print('Recall: %f' % recall)
# f1: tp / (tp + fp + fn)
f1 = f1_score(Y_test, pred_8)
print('F1 score: %f' % f1)
```

```python
# AUC,ROC Curve for Decision Tree: Max Depth 8

# predict probabilities
probs_8 = model_8.predict_proba(X_test)
# keep probabilities for the positive outcome only
probs_8 = probs_8[:, 1]

auc = roc_auc_score(Y_test, probs_8)
print('AUC - Test Set: %.2f%%' % (auc*100))

# calculate roc curve
fpr, tpr, thresholds = roc_curve(Y_test, probs_8)
# plot no skill
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
plt.xlabel('False positive rate')
plt.ylabel('Sensitivity/ Recall')
# show the plot
plt.show()
```

```python
# Random Forest

RF = RandomForestClassifier(criterion='gini', max_depth=None, max_features='auto',
                               max_leaf_nodes=None, bootstrap=True, oob_score=True, 
                               random_state=42, verbose=0)

RF.fit(X_train, Y_train)
```

```python
# Retrieve OOB Rate to understand the optimal number of decision trees needed

RF.set_params(warm_start=True, 
                  oob_score=True)

min_estimators = 100
max_estimators = 1000

error_rate = {}

for i in range(min_estimators, max_estimators + 1):
    RF.set_params(n_estimators=i)
    RF.fit(X_train, Y_train)

    oob_error = 1 - RF.oob_score_
    error_rate[i] = oob_error
    
# Convert dictionary to a pandas series for easy plotting 
oob_series = pd.Series(error_rate)
```

```python
fig, ax = plt.subplots(figsize=(20,10))

ax.set_facecolor('#fafafa')

oob_series.plot(kind='line',color = 'green')
plt.axhline(0.055, color='#875FDB',linestyle='--')
plt.axhline(0.05, color='#875FDB',linestyle='--')
plt.xlabel('n_estimators')
plt.ylabel('OOB Error Rate')
plt.title('OOB Error Rate Across various Forest sizes \n(From 100 to 1000 trees)')
```

```python
# From this we understand best 300 trees are the best number of estimators
```

```python
# Refine the tree via OOB Output taking 300 trees
RF.set_params(n_estimators=300,
                  bootstrap = True,
                  warm_start=False, 
                  oob_score=False)
```

```python
# Fitting the data
RF.fit(X_train,Y_train)
```

```python
# Performance and Accuracy Metrics for Random Forest with GridSearchCv

predictions_rf = RF.predict(X_test)
```

```python
# Confusion Matrix for Random forest
def create_conf_mat(test_class_set, predictions):
    """Function returns confusion matrix comparing two arrays"""
    if (len(test_class_set.shape) != len(predictions.shape) == 1):
        return print('Arrays entered are not 1-D.\nPlease enter the correctly sized sets.')
    elif (test_class_set.shape != predictions.shape):
        return print('Number of values inside the Arrays are not equal to each other.\nPlease make sure the array has the same number of instances.')
    else:
        # Set Metrics
        test_crosstb_comp = pd.crosstab(index = test_class_set,
                                        columns = predictions)
        # Changed for Future deprecation of as_matrix
        test_crosstb = test_crosstb_comp.values
        return test_crosstb

# Plotting Confusion Matrix
conf_mat = create_conf_mat(Y_test, predictions_rf)
sns.heatmap(conf_mat, annot=True, fmt='d', cbar=False)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Actual vs. Predicted Confusion Matrix')
plt.show()
```

```python
# Classification Report for Random Forest

clmrf = classification_report(Y_test, predictions_rf)
print(clmrf)
```

```python
# AUC,ROC Curve for Random Forest

# predict probabilities
probs_RF = RF.predict_proba(X_test)
# keep probabilities for the positive outcome only
probs_RF = probs_RF[:, 1]

auc = roc_auc_score(Y_test, probs_RF)
print('AUC - Test Set: %.2f%%' % (auc*100))

# calculate roc curve
fpr, tpr, thresholds = roc_curve(Y_test, probs_RF)
# plot no skill
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
plt.xlabel('False positive rate')
plt.ylabel('Sensitivity/ Recall')
# show the plot
plt.show()
```

```python
# K nearest Neighbours

# Detection of Optimal Value for K Neighnours
err = []

# Calculating error for K values between 1 and 50
for i in range(1, 50):  
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, Y_train)
    pred_i = knn.predict(X_test)
    err.append(np.mean(pred_i != Y_test))

# Plotting the error for K values between 1 and 50

plt.figure(figsize=(12, 6))  
plt.plot(range(1, 50), err, color='green', linestyle='dashed', marker='o',  
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')  
plt.xlabel('K Value')  
plt.ylabel('Mean Error')  
plt.show()
```
